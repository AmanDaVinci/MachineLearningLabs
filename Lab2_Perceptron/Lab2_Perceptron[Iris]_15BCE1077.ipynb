## Perceptron

```python
class Perceptron(object):
    """Perceptron classifier.

    Parameters
    ------------
    eta : float
        Learning rate (between 0.0 and 1.0)
    n_iter : int
        Passes over the training dataset.

    Attributes
    -----------
    w_ : 1d-array
        Weights after fitting.
    errors_ : list
        Number of misclassifications in every epoch.

    """
    def __init__(self, eta=0.01, n_iter=10):
        self.eta = eta
        self.n_iter = n_iter

    def fit(self, X, y):
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
            Training vectors, where n_samples is the number of samples and
            n_features is the number of features.
        y : array-like, shape = [n_samples]
            Target values.

        Returns
        -------
        self : object

        """
        self.w_ = np.zeros(1 + X.shape[1])
        self.errors_ = []

        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        """Compute linear activation"""
        return self.net_input(X)

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.activation(X) >= 0.0, 1, -1)
```

## Binary Target Selection

```python
# Dictionary to hold the binary targets
y = {}

y['setosa_versicolor'] = df.iloc[0:100,4].values
y['setosa_versicolor'] = np.where(y['setosa_versicolor'] == 'Iris-setosa', -1, 1)

y['versicolor_virginica'] = df.iloc[50:150,4].values
y['versicolor_virginica'] = np.where(y['versicolor_virginica'] == 'Iris-versicolor', -1, 1)

y['virginica_setosa'] = np.concatenate((df.iloc[100:150,4].values,df.iloc[0:50,4].values))
y['virginica_setosa'] = np.where(y['virginica_setosa'] == 'Iris-virginica', -1, 1)
```

## Feature Selection

```python
# Dictionary to hold the features
X = {}

X['setosa_versicolor'] = df.iloc[0:100, [0,1,2,3]].values
X['versicolor_virginica'] = df.iloc[50:150, [0,1,2,3]].values
X['virginica_setosa'] = np.vstack((df.iloc[100:150, [0,1,2,3]].values, df.iloc[0:50, [0,1,2,3]].values))
```

## Feature Engineering

**Selecting two features such that the classes are linearly separable**

```python
def visualize(X, class1, class2):
    '''Visualising features in 2D Space'''    
    features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
    combinations = [[0,1], [1,2], [2,3], [3,0]]
    for j in combinations:        
        t1_x0 = X[0:50][:,j[0]]
        t1_x1 = X[0:50][:,j[1]]
        t2_x0 = X[50:100][:,j[0]]
        t2_x1 = X[50:100][:,j[1]]

        ax = plt.figure(figsize=(5,4))
        plt.scatter(t1_x0, t1_x1, color='red', marker='o', label=class1)
        plt.scatter(t2_x0, t2_x1, color='blue', marker='x', label=class2)

        plt.xlabel(features[j[0]])
        plt.ylabel(features[j[1]])
        plt.legend(loc='upper left')
```

```python
visualize(X['setosa_versicolor'], 'setosa', 'versicolor')
```

```python
visualize(X['virginica_setosa'], 'versicolor', 'virginica')
```

```python
visualize(X['versicolor_virginica'], 'versicolor', 'virginica')
```

### From the visualizations one can conclude that:
- Setosa is linearly separable from both classes of Versicolor and Virginica.
- Versicolor and Virginica are only linearly separable with features : **Petal Width and Petal Length**.

### Linearly Separable Features :  Petal Width & Petal Length

```python
X['setosa_versicolor'] = X['setosa_versicolor'][:,(2,3)]
X['versicolor_virginica'] = X['versicolor_virginica'][:,(2,3)]
X['virginica_setosa'] = X['virginica_setosa'][:,(2,3)]
```

## Train the perceptrons

```python
# Dictionary to hold perceptron classifiers
clf = {}

clf['setosa_versicolor'] = Perceptron(eta=0.01, n_iter=1000)
clf['versicolor_virginica'] = Perceptron(eta=0.01, n_iter=1000)
clf['virginica_setosa'] = Perceptron(eta=0.01, n_iter=1000)
```

```python
clf['setosa_versicolor'].fit(X['setosa_versicolor'], y['setosa_versicolor'])
clf['versicolor_virginica'].fit(X['versicolor_virginica'], y['versicolor_virginica'])
clf['virginica_setosa'].fit(X['virginica_setosa'], y['virginica_setosa'])
```

## Plot the training session

```python
for key, H in clf.items():
    plt.plot(range(1, len(H.errors_) + 1), H.errors_, marker='.')
    plt.xlabel('Epochs')
    plt.ylabel('Misclassifications')
    plt.title(key)
    plt.show()
```

## Visualizing the Perceptron

```python
from matplotlib.colors import ListedColormap
```

```python
def plot_decision_regions(X, y, classifier, resolution=0.02):
    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                         np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=cmap(idx),
                    marker=markers[idx], label=cl)
```

```python
for key, H in clf.items(): 
    plot_decision_regions(X[key], y[key], classifier=H)
    plt.xlabel('petal width [cm]')
    plt.ylabel('petal length [cm]')
    plt.legend(loc='upper left')

    plt.tight_layout()
    plt.show()
```
