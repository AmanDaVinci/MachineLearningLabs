### One Hot Encode Categorical Features

```python
for i in cat_features:
    unique = len(data[i].unique())
    print("{}: {}".format(i, unique))
```

```python
for feature in cat_features:
    data = pd.get_dummies(data, prefix=feature, columns=[feature])
```
